```python
import os
import pandas as pd
import numpy as np

data_dir_path = os.path.join('train_data')

tsv_file_paths = [
    os.path.join(data_dir_path, 'Abl.tsv'),
    os.path.join(data_dir_path, 'Akt.tsv'),
    os.path.join(data_dir_path, 'Akt2.tsv'),
    os.path.join(data_dir_path, 'Akt3.tsv'),
    os.path.join(data_dir_path, 'ALK.tsv'),
    os.path.join(data_dir_path, 'ATM.tsv'),
    os.path.join(data_dir_path, 'b-Raf.tsv'),
    os.path.join(data_dir_path, 'b-Raf2.tsv'),
    os.path.join(data_dir_path, 'CDKs-1.tsv'),
    os.path.join(data_dir_path, 'CDKs-2.tsv'),
    os.path.join(data_dir_path, 'CDKs-5.tsv'),
    os.path.join(data_dir_path, 'CDKs-7.tsv'),
    os.path.join(data_dir_path, 'CDKs-9.tsv'),
    os.path.join(data_dir_path, 'Interleukin-1 receptor-associated kinase 1.tsv'),
    os.path.join(data_dir_path, 'Interleukin-1 receptor-associated kinase 3.tsv'),
    os.path.join(data_dir_path, 'Interleukin-1 receptor-associated kinase 4.tsv'),
    os.path.join(data_dir_path, 'JAK.tsv'),
    os.path.join(data_dir_path, 'JAK2.tsv'),
    os.path.join(data_dir_path, 'Rho kinase.tsv'),
    os.path.join(data_dir_path, 'ROCK2.tsv'),
    os.path.join(data_dir_path, 'Serinethreonine-protein kinase MST2.tsv'),
    os.path.join(data_dir_path, 'Thymidine kinase, cytosolic.tsv'),
    os.path.join(data_dir_path, 'Vascular endothelial growth factor receptor.tsv'),
    os.path.join(data_dir_path, 'Vascular endothelial growth factor receptor 2.tsv')
]

# CSV 파일 읽기 및 열 이름 추출
csv_data = pd.read_csv(os.path.join(data_dir_path, 'train.csv'))
csv_columns = csv_data.columns.tolist()

# 딕셔너리로 조정된 데이터프레임 저장
adjusted_dfs = []
for path in tsv_file_paths:
    filename = os.path.basename(path).split('.')[0]  # 파일 이름만 추출
    tsv_data = pd.read_csv(path, delimiter='\t')
    
    # 필요한 모든 열이 있는지 확인하고 없는 경우 NaN으로 채움
    for col in csv_columns:
        if col not in tsv_data.columns:
            tsv_data[col] = np.nan
    
    # train.csv의 열 순서에 맞게 열 정렬
    tsv_data = tsv_data[csv_columns]
    
    adjusted_dfs.append(tsv_data)

# 모든 데이터 프레임을 하나로 합치기
merged_data = pd.concat([csv_data] + adjusted_dfs, ignore_index=True)

print("Merged DataFrame shape:", merged_data.shape)
```

```python
merged_data.to_csv('train_data.csv', index=False)
```
